## Statistics package scipy.stats
https://docs.scipy.org/doc/scipy/reference/stats.html

```python
from scipy import stats
```

Example: Gaussian normal distribution (probability density function)

```python
x = np.linspace(0,10,100)
y = stats.norm.pdf(x,5,1.0) # x, mu, sigma
plt.plot(x,y)
plt.show()
```

cumulative density function

```python
y = stats.norm.cdf(x,5,1.0)
plt.plot(x,y)
plt.show()
```

inverse cumulative density function

```python
X = np.linspace(0,1,1000)
Y = stats.norm.ppf(X,0.5,0.3)
plt.plot(X,Y)
plt.show()
```

find Gaussian of a given distribution

```python
n= 500
y = np.random.randn(n)
plt.figure(0)
plt.plot(y)
# computing higher order moments (skewness and kurtosis)
print(np.mean(y),np.std(y),stats.skew(y),stats.kurtosis(y))

#histogram:
plt.figure(1)
bins = np.ceil(np.sqrt(n)*2).astype(np.int)
histX = np.linspace(-4,4,bins)

histo = np.histogram(y,bins,range=(-4,4))[0]
plt.plot(histX,histo,drawstyle='steps-mid')
plt.fill(histX, histo, 'b', alpha=0.4)

histY = stats.norm.pdf(histX,np.mean(y),np.std(y))
histY = histY/sum(histY)*n 
plt.plot(histX,histY)

plt.show()
```

## Fourier transform with scipy.fftpack
https://docs.scipy.org/doc/scipy/reference/fftpack.html

```python
import imageio
from scipy import fftpack
```

```python
im = np.clip(imageio.imread('imageio:camera.png')/255.,0.,1.)

ft = fftpack.fftshift(fftpack.fft2(im))
mag = np.abs(ft)
phase = np.angle(ft)
plt.figure(0)
plt.imshow(im,'gray')
plt.figure(1)
plt.imshow(np.log(mag+1),'gray')
plt.figure(2)
plt.imshow(phase,'jet')
plt.show()
```

Back-transform with only coarser frequencies

```python
mag2 = np.zeros(im.shape)

center = np.array(im.shape)//2

offset = (0.2 * center[0]).astype(np.int)
mag2[center[0]-offset:center[0]+offset,center[1]-offset:center[1]+offset] = 1
mag2 *= mag

# retransform
ft = mag2 * (np.cos(phase)+np.sin(phase)*1j)
imBack = np.abs(fftpack.ifft2(fftpack.ifftshift(ft)))

plt.figure(0)
epsilon= 0.00001
plt.imshow(np.log(mag2+epsilon),'gray')

plt.figure(1)
plt.imshow(imBack,'gray')
plt.show()
```

## Morphology with scipy.ndimage (you are doing this by hand once)
https://docs.scipy.org/doc/scipy-0.14.0/reference/ndimage.html

```python
from scipy import ndimage
```

```python
im = np.clip(imageio.imread('lena128C.png')/255.,0.,1.)
imgray = (im[:,:,0]+im[:,:,1]+im[:,:,2])/3.

plt.figure(0)
plt.imshow(imgray,'gray')

#mask
mask = np.ones((1,5)).astype(np.int)
dilated = ndimage.grey_dilation(imgray,footprint=mask)

plt.figure(1)
plt.imshow(dilated,'gray')
plt.show()
```

## Convolution with scipy.ndimage

```python
im = np.clip(imageio.imread('lena128C.png')/255.,0.,1.)
imgray = (im[:,:,0]+im[:,:,1]+im[:,:,2])/3.

plt.figure(0)
plt.imshow(imgray,'gray')

# gaussian mask
offset,b = 1,2
vx,vy = np.mgrid[-b:b+offset:offset,-b:b+offset:offset]
dist = (vx**2+vy**2)**0.5
gauss = stats.norm.pdf(dist)
gauss = gauss / np.sum(gauss)

plt.figure(1)
plt.imshow(gauss,'gray')

convolved = ndimage.convolve(imgray,gauss)

plt.figure(2)
plt.imshow(convolved,'gray')
plt.show()
```
